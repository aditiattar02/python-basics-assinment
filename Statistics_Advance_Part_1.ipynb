#Statistics Advance Part 1 (Theory)

Q1.What is a random variable in probability theory?
  - In probability theory, a random variable is a variable whose value is a numerical outcome of a random phenomenon, where a random phenomenon is one where the outcome is not predictable in advance, but can be described by a probability distribution.

Q2. What are the types of random variables?
  - Random variables are classified as either discrete (taking on a finite or countably infinite number of values) or continuous (taking on any value within a given range).

Q3. What is the difference between discrete and continuous distributions?
  - The key difference between discrete and continuous distributions lies in the nature of the values a random variable can take: discrete distributions have a countable, finite or infinite number of possible values, while continuous distributions can take any value within a specified range.

Q4. What are probability distribution functions (PDF)?
  - A probability distribution function (PDF) describes the likelihood of different possible values of a random variable, providing a mathematical function that assigns probabilities to events or outcomes.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - While the Probability Distribution Function (PDF) describes the probability of a random variable taking a specific value, the Cumulative Distribution Function (CDF) describes the probability of a random variable being less than or equal to a specific value.

Q6. What is a discrete uniform distribution?
  - a discrete uniform distribution is "a known, finite number of outcomes all equally likely to happen." A simple example of the discrete uniform distribution comes from throwing a fair six-sided die.

Q7. What are the key properties of a Bernoulli distribution?
  - Binary Outcomes:A Bernoulli random variable can only take on two values: 1 (success) or 0 (failure).

- Constant Probability of Success:The probability of success (p) is the same for each trial, and the probability of failure is 1-p.

- Independent Trials:
Each trial is independent of the others, meaning the outcome of one trial doesn't influence the outcome of any other trial.

- Discrete Probability Distribution:
The Bernoulli distribution is a discrete probability distribution, meaning the possible values are distinct and countable.

Q8. What is the binomial distribution, and how is it used in probability?
  - The binomial distribution is a discrete probability distribution that describes the probability of a certain number of successes in a fixed number of independent trials, each with the same probability of success. It's used to model situations with two possible outcomes (success or failure) and is a core concept in probability and statistics.

Q9. What is the Poisson distribution and where is it applied?
  - The Poisson distribution is a discrete probability distribution that models the probability of a certain number of events occurring within a fixed interval of time or space, given a known average rate of events. It's applied in scenarios with random, independent events occurring at a constant rate, like modeling customer arrivals, call center calls, or radioactive decay.

Q10. What is a continuous uniform distribution?
  - The uniform distribution (continuous) is one of the simplest probability distributions in statistics. It is a continuous distribution, this means that it takes values within a specified range, e.g. between 0 and 1.

Q11. What are the characteristics of a normal distribution?
  - A normal distribution, also known as a Gaussian distribution, is characterized by being symmetrical, bell-shaped, unimodal (having one peak), and having the mean, median, and mode all equAL.

Q12. What is the standard normal distribution, and why is it important?
  - The standard normal distribution is a specific type of normal distribution with a mean of 0 and a standard deviation of 1, crucial for simplifying probability calculations and comparing datasets.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) is a fundamental statistical concept stating that the distribution of sample means will approximate a normal distribution as the sample size increases, regardless of the original population distribution's shape. This is critical because it allows for statistical inferences and analysis using the normal distribution, even when the underlying data isn't normally distributed.

Q14. How does the Central Limit Theorem relate to the normal distribution?
  - The Central Limit Theorem (CLT) establishes that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the original population's distribution, making the normal distribution a cornerstone of statistical inference.

Q15. What is the application of Z statistics in hypothesis testing?
  - In hypothesis testing, Z-statistics (or Z-tests) are used to determine if there's a statistically significant difference between a sample mean and a population mean, or between the means of two groups, when the population standard deviation is known or the sample size is large.

Q16. How do you calculate a Z-score, and what does it represent?
  - A z-score, calculated as (x - μ) / σ, represents how many standard deviations a data point (x) is away from the mean (μ) of a distribution, with σ being the standard deviation.

Q17. What are point estimates and interval estimates in statistics?
  - In statistics, a point estimate is a single value used to estimate an unknown population parameter, while an interval estimate provides a range of values within which the parameter is likely to fall.

Q18. What is the significance of confidence intervals in statistical analysis?
  - Confidence intervals in statistical analysis provide a range of plausible values for an unknown population parameter, allowing researchers to estimate the true value with a certain level of confidence, while also quantifying the uncertainty in that estimate.

Q19. What is the relationship between a Z-score and a confidence intervaL?
  - A Z-score, which measures how many standard deviations a data point is from the mean, is directly related to the width of a confidence interval, with higher confidence levels (requiring larger Z-scores) resulting in wider intervals.

Q20. How are Z-scores used to compare different distributions?
  - Z-scores standardize data by expressing each value as the number of standard deviations it deviates from the mean, enabling meaningful comparisons across different distributions, even those with varying means and standard deviations.

Q21. What are the assumptions for applying the Central Limit Theorem?
  - To apply the Central Limit Theorem (CLT), you need to assume that the samples are independent, identically distributed (i.i.d.) random variables, and that the population has a finite variance. Additionally, a sufficiently large sample size is generally required.

Q22. What is the concept of expected value in a probability distribution?
  - n a probability distribution, the expected value (also known as the mean or mathematical expectation) is the theoretical average outcome of a random variable if the experiment is repeated many times, calculated by weighting each possible outcome by its probability.

Q23. How does a probability distribution relate to the expected outcome of a random variable?
  - A probability distribution describes the likelihood of different outcomes for a random variable, and the expected outcome is calculated by averaging each possible outcome weighted by its probability.

```python
#Q6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
'''
ANSWER
'''
import numpy as np
import matplotlib.pyplot as plt

low = 1
high = 6

x_values = np.arange(low, high + 1)

cdf_values = (x_values - low + 1) / (high - low + 1)

plt.figure(figsize=(8, 6))
plt.step(x_values, cdf_values, where='post', color='GREEN', linestyle='-', linewidth=2, marker='o', markersize=8)
plt.title('Cumulative Distribution Function (CDF) of a Discrete Uniform Distribution')
plt.xlabel('Outcomes')
plt.ylabel('Cumulative Probability')
plt.xticks(np.arange(low, high + 1))
plt.yticks(np.arange(0, 1.1, 0.2))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q9. Write a Python function to calculate Z-scores from a dataset and plot them.
'''
ANSWER
'''
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

data = np.random.normal(0, 1, 1000)

z_scores = calculate_z_scores(data)

plt.figure(figsize=(8, 6))
plt.plot(z_scores, color='CYAN', marker='o', markersize=4, linestyle='-', alpha=0.7)
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.axhline(y=0, color='r', linestyle='--', label='Mean (Z=0)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q16. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
'''
ANSWER
'''
import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal_distribution():
    x = np.linspace(-4, 4, 1000)

    pdf_values = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf_values, color='blue', label='Standard Normal Distribution (mean=0, std=1)')
    plt.fill_between(x, pdf_values, color='PINK', alpha=0.5)
    plt.title('Standard Normal Distribution (mean=0, std=1)')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

plot_standard_normal_distribution()
```

```python
#Q19. Implement hypothesis testing using Z-statistics for a sample dataset.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

sample_data = np.array([72, 75, 80, 65, 85, 78, 68, 72, 74, 81,
                        69, 71, 77, 74, 79, 73, 68, 70, 76, 82,
                        69, 75, 70, 72, 78, 74, 71, 69, 73, 80])

mu = 70
sigma = 10
n = len(sample_data)
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

z_statistic = (sample_mean - mu) / (sigma / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

if p_value < alpha:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")
print(f"Result: {result}")

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Standard Normal Distribution")
plt.fill_between(x, y, where=(x >= z_statistic), color='skyblue', alpha=0.5, label=f"Z-statistic region (Z={z_statistic:.2f})")
plt.axvline(z_statistic, color='red', linestyle='dashed', label=f"Z-statistic ({z_statistic:.2f})")
plt.title("Z-Test Hypothesis Testing")
plt.xlabel("Z-value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q20. Create a confidence interval for a dataset using Python and interpret the result.
'''
ANSWER'''

import numpy as np
import scipy.stats as stats

# Sample data (e.g., exam scores)
sample_data = np.array([72, 75, 80, 65, 85, 78, 68, 72, 74, 81,
                        69, 71, 77, 74, 79, 73, 68, 70, 76, 82,
                        69, 75, 70, 72, 78, 74, 71, 69, 73, 80])

confidence_level = 0.95
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
alpha = 1 - confidence_level

t_critical = stats.t.ppf(1 - alpha / 2, df=n - 1)

margin_of_error = t_critical * (sample_std / np.sqrt(n))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"T-critical value: {t_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"Confidence Interval: {confidence_interval}")
```

```python
#Q21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

np.random.seed(42)
mean = 50
std_dev = 10
sample_size = 100

sample_data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

std_error = sample_std / np.sqrt(sample_size)

confidence_level = 0.95
alpha = 1 - confidence_level

confidence_interval = stats.t.interval(confidence_level, df=sample_size-1, loc=sample_mean, scale=std_error)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Confidence Interval: {confidence_interval}")
```

```python
#Q27. Implement Z-score normalization on a dataset and visualize its transformation.
'''
ANSWER
'''
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
mean = 50
std_dev = 10
sample_size = 1000

dataset = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

mean_data = np.mean(dataset)
std_dev_data = np.std(dataset)
z_scores = (dataset - mean_data) / std_dev_data

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(dataset, bins=30, color='GREEN', edgecolor='black')
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Dataset')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#Q28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

np.random.seed(42)
mean = 50
std_dev = 10
sample_size = 1000

dataset = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

skewness = stats.skew(dataset)
kurtosis = stats.kurtosis(dataset)

print(f"Skewness of the dataset: {skewness}")
print(f"Kurtosis of the dataset: {kurtosis}")
```

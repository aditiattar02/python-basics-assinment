 Q2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.
   - Measures of central tendency are statistical measures that describe the center or typical value of a set of data. The three most commonly used measures of central tendency are mean, median, and mode. Each measure is useful in different situations depending on the nature of the data and the context of the analysis.

1. Mean
The mean is the arithmetic average of a set of numbers. It is calculated by adding up all the values in a dataset and then dividing by the number of values.

 example: Consider the following dataset: 2, 4, 6, 8, 10. The mean would be:
  2+4+6+8+10 divided by 5 = 6

2. Median
The median is the middle value in a dataset when the values are arranged in order (from least to greatest or greatest to least). If the dataset has an odd number of values, the median is the middle one; if the dataset has an even number of values, the median is the average of the two middle values.

Example:
Consider the following dataset: 2, 4, 6, 8, 10. Since there are 5 numbers, the median is the middle one, which is 6.

Now consider this dataset: 2, 4, 6, 8. The median is the average of 4 and 6:
4+6 / 2 = 5

3. Mode
The mode is the value that occurs most frequently in a dataset. A dataset can have:

One mode (unimodal),
Two modes (bimodal),
More than two modes (multimodal),
Or no mode at all if no number repeats.
Example:
Consider the following dataset: 1, 2, 2, 3, 4, 5, 5, 5. The mode is 5 because it appears more frequently than any other number.

Q3.  Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

  - Dispersion refers to the extent to which data values in a dataset are spread out or scattered. It gives us an idea of how much individual data points differ from the central tendency (mean, median, or mode). Understanding dispersion is essential because it provides insights into the variability or consistency of the data. In other words, while measures of central tendency (like the mean or median) tell us about the "center" of the data, measures of dispersion tell us about the "spread."

Two of the most commonly used measures of dispersion are variance and standard deviation.

1. Variance
Variance measures how far each data point in a dataset is from the mean and, consequently, how spread out the data points are. It is essentially the average of the squared differences from the mean. By squaring the differences, variance ensures that both positive and negative deviations contribute equally to the overall spread.

2. Standard Deviation
Standard deviation is the square root of the variance. It provides a measure of spread in the same units as the original data, making it more interpretable than variance. It is the most widely used measure of dispersion and gives a clear indication of how much the data deviates from the mean.

Q12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.

 - Letâ€™s go through an example dataset and calculate **covariance** and **correlation** to help you understand how these two measures work.

### Example Dataset:

Suppose we have the following dataset of two variables: **X** and **Y**. These could represent any two related measurements, such as the number of hours studied (X) and the scores on a test (Y).

| X (Hours Studied) | Y (Test Score) |
|-------------------|----------------|
| 1                 | 50             |
| 2                 | 55             |
| 3                 | 60             |
| 4                 | 65             |
| 5                 | 70             |

### Step 1: **Calculate the Mean of X and Y**

First, calculate the mean of each variable:

\[
\text{Mean of X} (\mu_X) = \frac{1 + 2 + 3 + 4 + 5}{5} = 3
\]
\[
\text{Mean of Y} (\mu_Y) = \frac{50 + 55 + 60 + 65 + 70}{5} = 60
\]

### Step 2: **Calculate Covariance**

The formula for **covariance** is:

\[
\text{Cov}(X, Y) = \frac{\sum (x_i - \mu_X)(y_i - \mu_Y)}{n}
\]

Where:
- \( x_i \) and \( y_i \) are individual data points of X and Y,
- \( \mu_X \) and \( \mu_Y \) are the means of X and Y,
- \( n \) is the number of data points.

Now, we calculate the covariance for this dataset:

| X | Y  | \( x_i - \mu_X \) | \( y_i - \mu_Y \) | \( (x_i - \mu_X)(y_i - \mu_Y) \) |
|---|----|-------------------|-------------------|----------------------------------|
| 1 | 50 | -2                | -10               | 20                               |
| 2 | 55 | -1                | -5                | 5                                |
| 3 | 60 | 0                 | 0                 | 0                                |
| 4 | 65 | 1                 | 5                 | 5                                |
| 5 | 70 | 2                 | 10                | 20                               |

Sum of \( (x_i - \mu_X)(y_i - \mu_Y) \) = \( 20 + 5 + 0 + 5 + 20 = 50 \)

Now, calculate the covariance:
\[
\text{Cov}(X, Y) = \frac{50}{5} = 10
\]

### Step 3: **Calculate Correlation**

The formula for **correlation** is:

\[
\text{Correlation} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
\]

Where:
- \( \sigma_X \) is the standard deviation of X,
- \( \sigma_Y \) is the standard deviation of Y.

To calculate the standard deviations, we first calculate the variances.

**Variance of X**:
\[
\text{Var}(X) = \frac{\sum (x_i - \mu_X)^2}{n} = \frac{4 + 1 + 0 + 1 + 4}{5} = 2
\]
\[
\sigma_X = \sqrt{2} \approx 1.41
\]

**Variance of Y**:
\[
\text{Var}(Y) = \frac{\sum (y_i - \mu_Y)^2}{n} = \frac{100 + 25 + 0 + 25 + 100}{5} = 50
\]
\[
\sigma_Y = \sqrt{50} \approx 7.07
\]

Now, we can calculate the correlation:
\[
\text{Correlation} = \frac{10}{1.41 \times 7.07} \approx \frac{10}{9.97} \approx 1
\]

### Step 4: **Interpretation**

- **Covariance**: The covariance of 10 indicates that as X increases, Y also tends to increase. A positive covariance means that both variables move in the same direction.
  
- **Correlation**: The correlation of approximately **1** suggests a **perfect positive linear relationship** between X and Y. This means that for every unit increase in X, Y increases in a perfectly predictable manner, and the relationship is very strong.

#STATISTICS ADVANCE PART 2 (THEORY)

Q1. What is hypothesis testing in statisticS?
  - Statistical hypothesis testing is a method used to determine if data supports a specific claim or hypothesis by analyzing a sample and comparing it to a null hypothesis, which assumes no effect or difference.

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - n hypothesis testing, the null hypothesis (H0) proposes no effect or difference, while the alternative hypothesis (H1 or Ha) suggests a real effect or difference that the researcher is trying to prove.

Q3. What is the significance level in hypothesis testing, and why is it important?
  -In hypothesis testing, the significance level (often denoted as α) is the probability of rejecting the null hypothesis when it is actually true, and it's crucial for determining the threshold at which a result is considered statistically significant.

Q4. What does a P-value represent in hypothesis testing?
  - In hypothesis testing, a p-value represents the probability of observing results as extreme as, or more extreme than, the ones obtained, assuming the null hypothesis is true.

Q5. How do you interpret the P-value in hypothesis testing?
  - In hypothesis testing, the p-value indicates the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. A small p-value suggests strong evidence against the null hypothesis, leading to its rejection.

Q6. What are Type 1 and Type 2 errors in hypothesis testing?
  - In hypothesis testing, a Type I error (false positive) occurs when you reject a true null hypothesis, while a Type II error (false negative) occurs when you fail to reject a false null hypothesis.

Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - In hypothesis testing, a one-tailed test checks for a difference in one specific direction (e.g., greater than or less than), while a two-tailed test checks for a difference in either direction (greater than or less than).

Q8.What is the Z-test, and when is it used in hypothesis testing?
  - A z-test is a statistical test used in hypothesis testing to determine if there's a significant difference between a sample mean and a population mean, or between two population means, when the population standard deviation is known, and the sample size is large (typically greater than 30).

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - To calculate a z-score, you subtract the population mean from a raw score and divide the result by the population standard deviation. In hypothesis testing, a z-score indicates how many standard deviations a data point is from the mean, helping determine if a sample result is statistically significant.

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The t-distribution is used as an alternative to the normal distribution when dealing with small sample sizes and unknown population standard deviations, while the normal distribution is used when the population standard deviation is known or the sample size is large.

Q11. What is the difference between a Z-test and a T-test?
  - The primary difference between a Z-test and a T-test lies in the sample size and knowledge of the population standard deviation. Z-tests are used with large sample sizes (n > 30) and known population standard deviation, while T-tests are used with smaller sample sizes (n < 30) or when the population standard deviation is unknown.

Q12. What is the T-test, and how is it used in hypothesis testing?
  - A t-test is a statistical test used to determine if there's a significant difference between the means of two groups, and it's a crucial tool in hypothesis testing when comparing averages and assessing the significance of differences.

Q13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Z-tests and t-tests are both used in hypothesis testing to compare means, but they differ primarily in their assumptions about sample size and population standard deviation: Z-tests are used for large samples (n > 30) and known population standard deviation, while t-tests are for smaller samples (n < 30) or unknown population standard deviation.

Q14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval (CI) is a range of values used to estimate an unknown population parameter (like a mean) based on sample data, with a certain level of confidence (e.g., 95%) that the true value lies within that range. It's used to quantify uncertainty and interpret statistical results by providing a plausible range for the true value.

Q15. What is the margin of error, and how does it affect the confidence interval?
  - The margin of error quantifies the potential inaccuracy of a sample statistic as an estimate of a population parameter, with a confidence interval being the range within which the true population value is likely to fall, and the margin of error is half the width of that interval.

Q16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes' Theorem, a fundamental concept in probability and statistics, allows you to update the probability of a hypothesis based on new evidence, moving from a prior probability to a posterior probability. Its significance lies in its ability to incorporate prior knowledge and new data to make more informed predictions and decisions under uncertainty.

Q17. What is the Chi-square distribution, and when is it used?
  - The chi-square distribution is a continuous probability distribution used in hypothesis testing, particularly for categorical data, to determine if observed data significantly differs from expected data, or if two categorical variables are related.

Q18. What is the Chi-square goodness of fit test, and how is it applied?
  - The chi-square goodness-of-fit test determines how well a sample data distribution fits a hypothesized distribution, by comparing observed frequencies to expected frequencies. It's a non-parametric test used to assess the "goodness of fit" of a statistical model to a set of observations.

Q19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is a continuous probability distribution used in hypothesis testing, particularly when comparing variances or means across multiple groups, and is commonly used in ANOVA (Analysis of Variance) and F-tests.

Q20. What is an ANOVA test, and what are its assumptions?
  - ANOVA (Analysis of Variance) is a statistical test used to compare the means of two or more groups, determining if there are statistically significant differences between them. Its key assumptions include independence of observations, normality of data within each group, and homogeneity of variance (equal variances across groups).

Q21. What are the different types of ANOVA tests?
  - The main types of ANOVA tests are one-way ANOVA (examines the effect of a single independent variable) and two-way ANOVA (evaluates the impact of two independent variables and their interaction).

Q22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test, a statistical test used in hypothesis testing, compares variances between two or more groups to determine if there's a significant difference in their means, often used in Analysis of Variance (ANOVA) and regression analysis.

```python
#Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest

population_mean = 100
population_std = 15

np.random.seed(42)
sample_size = 30
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

z_score, p_value = ztest(sample_data, value=population_mean, alternative='two-sided')

alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Standard Normal Distribution", color="blue")

plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label="Rejection Region")

plt.axvline(-critical_value, color='red', linestyle='dashed', label=f"Critical Value (-{critical_value:.2f})")
plt.axvline(critical_value, color='red', linestyle='dashed', label=f"Critical Value ({critical_value:.2f})")

plt.axvline(z_score, color='green', linestyle='solid', label=f"Observed Z-Score ({z_score:.2f})")

plt.title("Two-Tailed Z-Test: Decision Regions")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()


plt.show()

print(f"Sample Mean: {np.mean(sample_data):.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Critical Value: ±{critical_value:.4f}")

if abs(z_score) > critical_value:
    print("Reject the null hypothesis (H0): Sample mean significantly differs from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0): No significant difference.")
```

```python
#Q5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(pop_mean=100, pop_std=15, sample_size=30, effect_size=5, alpha=0.05):

    std_error = pop_std / np.sqrt(sample_size)


    x = np.linspace(pop_mean - 4*std_error, pop_mean + 4*std_error, 1000)
    null_dist = stats.norm.pdf(x, pop_mean, std_error)
    alt_dist = stats.norm.pdf(x, pop_mean + effect_size, std_error)


    critical_value = pop_mean + stats.norm.ppf(1 - alpha) * std_error


    beta = stats.norm.cdf(critical_value, loc=pop_mean + effect_size, scale=std_error)


    plt.figure(figsize=(9, 5))
    plt.plot(x, null_dist, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Hypothesis (H1)", color="green")

    plt.fill_between(x, null_dist, where=(x >= critical_value), color='red', alpha=0.3, label="Type 1 Error (α)")


    plt.fill_between(x, alt_dist, where=(x < critical_value), color='purple', alpha=0.3, label="Type 2 Error (β)")


    plt.axvline(critical_value, color='red', linestyle='dashed', label=f"Critical Value ({critical_value:.2f})")


    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()


    plt.show()


    print(f"Critical Value: {critical_value:.4f}")
    print(f"Type 1 Error (α): {alpha}")
    print(f"Type 2 Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")


visualize_errors()
```

```python
#Q6. Write a Python program to perform an independent T-test and interpret the results.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):

    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)


    print(f"Sample 1 Mean: {np.mean(sample1):.4f}, Sample 2 Mean: {np.mean(sample2):.4f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")


    if p_value < alpha:
        print("Reject the null hypothesis (H0): The means are significantly different.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference between means.")


np.random.seed(42)
group1 = np.random.normal(loc=100, scale=15, size=30)
group2 = np.random.normal(loc=110, scale=15, size=30)


independent_t_test(group1, group2)
```

```python
#Q7. Perform a paired sample T-test using Python and visualize the comparison results?
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(before, after, alpha=0.05):
    differences = np.array(after) - np.array(before)

    t_stat, p_value = stats.ttest_rel(before, after)

    print(f"Mean Before: {np.mean(before):.4f}, Mean After: {np.mean(after):.4f}")
    print(f"Mean Difference: {np.mean(differences):.4f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0): Significant difference before and after.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference.")

    plt.figure(figsize=(8, 5))

    plt.subplot(1, 2, 1)
    plt.boxplot([before, after], labels=["Before", "After"], patch_artist=True)
    plt.title("Before vs. After")
    plt.ylabel("Values")

    plt.subplot(1, 2, 2)
    plt.plot(before, after, 'o', markersize=6, color="gray", alpha=0.6, label="Individual Data Points")
    plt.plot([np.mean(before)], [np.mean(after)], 'ro', markersize=8, label="Mean Change")
    plt.plot([min(before), max(before)], [min(before), max(before)], 'k--', label="No Change Line")

    plt.xlabel("Before")
    plt.ylabel("After")
    plt.title("Mean Comparison")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

np.random.seed(42)
before_scores = np.random.normal(loc=75, scale=10, size=20)
after_scores = before_scores + np.random.normal(loc=3, scale=5, size=20)

paired_t_test(before_scores, after_scores)
```

```python
#Q8. Simulate data and perform both Z-test and T-test, then compare the results using Python.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest
import matplotlib.pyplot as plt

def simulate_and_compare_tests(sample_size=30, pop_mean=100, pop_std=15, alpha=0.05):
    np.random.seed(42)
    sample_data = np.random.normal(loc=pop_mean, scale=pop_std, size=sample_size)

    z_stat, z_p_value = ztest(sample_data, value=pop_mean, alternative='two-sided')

    t_stat, t_p_value = stats.ttest_1samp(sample_data, pop_mean)

    print(f"Sample Mean: {np.mean(sample_data):.4f}")
    print("\nZ-Test Results:")
    print(f"  Z-Statistic: {z_stat:.4f}")
    print(f"  P-Value: {z_p_value:.4f}")

    print("\nT-Test Results:")
    print(f"  T-Statistic: {t_stat:.4f}")
    print(f"  P-Value: {t_p_value:.4f}")

    def interpret_test(p_value, test_name):
        if p_value < alpha:
            print(f"  {test_name}: Reject the null hypothesis (H0) → Significant difference.")
        else:
            print(f"  {test_name}: Fail to reject the null hypothesis (H0) → No significant difference.")

    interpret_test(z_p_value, "Z-Test")
    interpret_test(t_p_value, "T-Test")

    x = np.linspace(-4, 4, 1000)
    normal_dist = stats.norm.pdf(x, 0, 1)
    t_dist = stats.t.pdf(x, df=sample_size - 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, normal_dist, label="Z-Test (Standard Normal)", color="blue")
    plt.plot(x, t_dist, label=f"T-Test (df={sample_size - 1})", color="green", linestyle="dashed")


    plt.axvline(z_stat, color='blue', linestyle='solid', label=f"Z-Statistic ({z_stat:.2f})")
    plt.axvline(t_stat, color='green', linestyle='solid', label=f"T-Statistic ({t_stat:.2f})")

    plt.title("Z-Test vs. T-Test: Distribution Comparison")
    plt.xlabel("Test Statistic Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()


    plt.show()


simulate_and_compare_tests()
```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_error = stats.sem(sample)

    if n < 30:
        t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
    else:
        t_value = stats.norm.ppf((1 + confidence) / 2)

    margin_of_error = t_value * std_error


    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=25)

ci_lower, ci_upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_error = stats.sem(sample)

    if n < 30:
        critical_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
    else:
        critical_value = stats.norm.ppf((1 + confidence) / 2)

    moe = critical_value * std_error
    return moe

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=25)

moe_value = margin_of_error(sample_data)
print(f"95% Margin of Error: ±{moe_value:.2f}")
```

```python
#Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
'''
ANSWER
'''
def bayes_theorem(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

prior_prob = 0.01
likelihood = 0.95
false_positive = 0.05
prob_no_disease = 1 - prior_prob

evidence_prob = (likelihood * prior_prob) + (false_positive * prob_no_disease)


posterior_prob = bayes_theorem(prior_prob, likelihood, evidence_prob)

print(f"Updated Probability of Having Disease Given a Positive Test: {posterior_prob:.4f}")
```

```python
#Q12. Perform a Chi-square test for independence between two categorical variables in Python.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import pandas as pd

def chi_square_test(data):
    chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)

    print(f"Chi-Square Statistic: {chi2_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Degrees of Freedom: {dof}")
    print("\nExpected Frequencies:")
    print(pd.DataFrame(expected, index=data.index, columns=data.columns))

    alpha = 0.05
    if p_value < alpha:
        print("\nReject the null hypothesis (H0): The variables are dependent.")
    else:
        print("\nFail to reject the null hypothesis (H0): The variables are independent.")

data = pd.DataFrame([[30, 10],
                     [20, 40]],
                   columns=["Liked", "Disliked"],
                   index=["Male", "Female"])

chi_square_test(data)
```

```python
#Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
'''
ANSWER
'''
import numpy as np
import pandas as pd
import scipy.stats as stats

def compute_expected_frequencies(observed_data):
    _, _, _, expected = stats.chi2_contingency(observed_data)

    expected_df = pd.DataFrame(expected, index=observed_data.index, columns=observed_data.columns)
    return expected_df

observed_data = pd.DataFrame([[30, 10],
                              [20, 40]],
                            columns=["Liked", "Disliked"],
                            index=["Male", "Female"])

expected_frequencies = compute_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies (If Independent):")
print(expected_frequencies)
```

```python
#Q14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed, expected):
    chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

    print(f"Chi-Square Statistic: {chi2_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print("\nReject the null hypothesis (H0): The observed data does NOT follow the expected distribution.")
    else:
        print("\nFail to reject the null hypothesis (H0): The observed data follows the expected distribution.")

observed_counts = [8, 10, 12, 14, 9, 7]
expected_counts = [10, 10, 10, 10, 10, 10]

goodness_of_fit_test(observed_counts, expected_counts)
```

```python
#Q16. Implement an F-test using Python to compare the variances of two random samples.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2, alpha=0.05):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        F_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        F_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    p_value = 2 * (1 - stats.f.cdf(F_stat, dfn, dfd))

    print(f"F-Statistic: {F_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print("\nReject the null hypothesis (H0): The variances are significantly different.")
    else:
        print("\nFail to reject the null hypothesis (H0): No significant difference in variances.")

np.random.seed(42)
sample1 = np.random.normal(loc=100, scale=15, size=30)
sample2 = np.random.normal(loc=100, scale=10, size=30)

f_test(sample1, sample2)
```

```python
#Q17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

def anova_test(*groups):
    F_stat, p_value = stats.f_oneway(*groups)

    print(f"F-Statistic: {F_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print("\nReject the null hypothesis (H0): At least one group mean is significantly different.")
    else:
        print("\nFail to reject the null hypothesis (H0): No significant difference between group means.")

np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=10, size=30)
group3 = np.random.normal(loc=85, scale=10, size=30)

anova_test(group1, group2, group3)
```

```python
#Q18. Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def anova_with_plot(*groups):
    F_stat, p_value = stats.f_oneway(*groups)

    print(f"F-Statistic: {F_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print("\nReject the null hypothesis (H0): At least one group mean is significantly different.")
    else:
        print("\nFail to reject the null hypothesis (H0): No significant difference between group means.")

    labels = []
    values = []
    for i, group in enumerate(groups):
        labels.extend([f"Group {i+1}"] * len(group))
        values.extend(group)

    df = pd.DataFrame({'Group': labels, 'Values': values})

    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Group', y='Values', data=df, palette="Set2")
    plt.title("One-Way ANOVA: Comparison of Group Means")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid(True)
    plt.show()

np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=10, size=30)
group3 = np.random.normal(loc=85, scale=10, size=30)

anova_with_plot(group1, group2, group3)
```

```python
#Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):
    alpha = 0.05

    print("\n**Normality Test (Shapiro-Wilk) for Each Group:**")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: W-statistic = {stat:.4f}, P-value = {p_value:.4f}")
        if p_value < alpha:
            print("  -> Not normally distributed ❌")
        else:
            print("  -> Normally distributed ✅")

    print("\n**Homogeneity of Variance Test (Levene’s Test):**")
    stat, p_value = stats.levene(*groups)
    print(f"Levene’s Test Statistic = {stat:.4f}, P-value = {p_value:.4f}")
    if p_value < alpha:
        print("  -> Variances are significantly different ❌ (violates assumption)")
    else:
        print("  -> Variances are equal ✅ (assumption met)")

    labels = []
    values = []
    for i, group in enumerate(groups):
        labels.extend([f"Group {i+1}"] * len(group))
        values.extend(group)

    plt.figure(figsize=(10, 5))
    for i, group in enumerate(groups):
        sns.histplot(group, kde=True, label=f'Group {i+1}', bins=15, alpha=0.6)
    plt.title("Histogram of Groups (Normality Check)")
    plt.legend()
    plt.grid()
    plt.show()

    plt.figure(figsize=(8, 5))
    sns.boxplot(x=labels, y=values, palette="Set2")
    plt.title("Boxplot of Groups (Variance Check)")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid()
    plt.show()

np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=10, size=30)
group3 = np.random.normal(loc=85, scale=12, size=30)

check_anova_assumptions(group1, group2, group3)
```

```python
#Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
'''
ANSWER
'''
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)
data = pd.DataFrame({
    'Teaching_Method': np.repeat(['Traditional', 'Online'], 30),
    'Study_Hours': np.tile(np.repeat(['Low', 'High'], 15), 2),
    'Exam_Score': np.concatenate([
        np.random.normal(70, 10, 15),
        np.random.normal(80, 10, 15),
        np.random.normal(72, 10, 15),
        np.random.normal(78, 10, 15)
    ])
})

model = smf.ols('Exam_Score ~ Teaching_Method + Study_Hours + Teaching_Method:Study_Hours', data=data).fit()
anova_results = sm.stats.anova_lm(model, typ=2)

print("\n**Two-Way ANOVA Results:**\n")
print(anova_results)

alpha = 0.05
for factor in anova_results.index:
    p_value = anova_results.loc[factor, 'PR(>F)']
    if p_value < alpha:
        print(f"\n{factor} has a significant effect on Exam Score (p={p_value:.4f}) ✅")
    else:
        print(f"\n{factor} does not have a significant effect on Exam Score (p={p_value:.4f}) ❌")

plt.figure(figsize=(8,6))
sns.pointplot(x="Study_Hours", y="Exam_Score", hue="Teaching_Method", data=data, capsize=0.1, markers=['o', 's'], linestyles=['-', '--'])
plt.title("Interaction Effect: Teaching Method vs. Study Hours on Exam Score")
plt.ylabel("Exam Score")
plt.grid()
plt.show()
```

```python
#Q21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_f_distribution(df1, df2):

    x = np.linspace(0, 5, 1000)
    f_dist = stats.f.pdf(x, df1, df2)

    alpha = 0.05
    f_critical = stats.f.ppf(1 - alpha, df1, df2)

    plt.figure(figsize=(8, 5))
    plt.plot(x, f_dist, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')

    x_crit = np.linspace(f_critical, 5, 500)
    plt.fill_between(x_crit, stats.f.pdf(x_crit, df1, df2), alpha=0.4, color='red', label="Critical Region (α=0.05)")

    plt.axvline(f_critical, linestyle='--', color='red', label=f'Critical F={f_critical:.2f}')
    plt.title("F-Distribution and Critical Region")
    plt.xlabel("F-Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

plot_f_distribution(df1=5, df2=20)
```

```python
#Q22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.
'''
ANSWER
'''
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=10, size=30)
group3 = np.random.normal(loc=85, scale=10, size=30)

data = pd.DataFrame({
    'Score': np.concatenate([group1, group2, group3]),
    'Group': ['Group 1'] * 30 + ['Group 2'] * 30 + ['Group 3'] * 30
})


F_stat, p_value = stats.f_oneway(group1, group2, group3)

print("\n**One-Way ANOVA Results:**")
print(f"F-Statistic: {F_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis (H0): At least one group mean is significantly different. ✅")
else:
    print("\nFail to reject the null hypothesis (H0): No significant difference between group means. ❌")

plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Score', data=data, palette="Set2")
plt.title("One-Way ANOVA: Comparison of Group Means")
plt.xlabel("Groups")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

```python
#Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=10, size=30)

t_stat, p_value = stats.ttest_ind(group1, group2)

print("\n**Two-Sample t-Test Results:**")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis (H0): The means are significantly different.")
else:
    print("\nFail to reject the null hypothesis (H0): No significant difference in means.")

plt.figure(figsize=(8, 5))
sns.histplot(group1, kde=True, label="Group 1", color="blue", alpha=0.6)
sns.histplot(group2, kde=True, label="Group 2", color="red", alpha=0.6)
plt.axvline(np.mean(group1), color='blue', linestyle='--', label=f"Mean 1: {np.mean(group1):.2f}")
plt.axvline(np.mean(group2), color='red', linestyle='--', label=f"Mean 2: {np.mean(group2):.2f}")
plt.title("Histogram of Two Groups")
plt.xlabel("Values")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
#Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample = np.random.normal(loc=75, scale=10, size=30)

sigma_0_squared = 100

n = len(sample)
sample_variance = np.var(sample, ddof=1)

chi_square_stat = (n - 1) * sample_variance / sigma_0_squared

alpha = 0.05
df = n - 1

chi_critical_low = stats.chi2.ppf(alpha / 2, df)
chi_critical_high = stats.chi2.ppf(1 - alpha / 2, df)

p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

print("\n**Chi-Square Test for Population Variance:**")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Test Statistic: {chi_square_stat:.4f}")
print(f"Critical Values: ({chi_critical_low:.4f}, {chi_critical_high:.4f})")
print(f"P-Value: {p_value:.4f}")

if chi_square_stat < chi_critical_low or chi_square_stat > chi_critical_high:
    print("\nReject the null hypothesis (H0): The population variance is significantly different.")
else:
    print("\nFail to reject the null hypothesis (H0): No significant difference in variance.")
```

```python
#Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats

def z_test_proportions(x1, n1, x2, n2, alpha=0.05):

    p1 = x1 / n1
    p2 = x2 / n2

    p_pooled = (x1 + x2) / (n1 + n2)

    se = np.sqrt(p_pooled * (1 - p_pooled) * (1/n1 + 1/n2))

    z_score = (p1 - p2) / se

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    z_critical = stats.norm.ppf(1 - alpha/2)

    print("\n**Two-Proportion Z-Test Results:**")
    print(f"Sample Proportion 1: {p1:.4f}")
    print(f"Sample Proportion 2: {p2:.4f}")
    print(f"Z-Score: {z_score:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Critical Z-Value (α={alpha}): ±{z_critical:.4f}")

    if abs(z_score) > z_critical:
        print("\nReject the null hypothesis (H0): The proportions are significantly different.")
    else:
        print("\nFail to reject the null hypothesis (H0): No significant difference in proportions.")

z_test_proportions(x1=60, n1=200, x2=75, n2=220)
```

```python
#Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def f_test_variance(sample1, sample2, alpha=0.05):
    # Sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        F = var1 / var2
        df1, df2 = len(sample1) - 1, len(sample2) - 1
    else:
        F = var2 / var1
        df1, df2 = len(sample2) - 1, len(sample1) - 1

    p_value = 2 * min(stats.f.cdf(F, df1, df2), 1 - stats.f.cdf(F, df1, df2))

    F_critical_low = stats.f.ppf(alpha / 2, df1, df2)
    F_critical_high = stats.f.ppf(1 - alpha / 2, df1, df2)

    print("\n**F-Test for Equality of Variances:**")
    print(f"Sample Variance 1: {var1:.4f}")
    print(f"Sample Variance 2: {var2:.4f}")
    print(f"F-Statistic: {F:.4f}")
    print(f"Degrees of Freedom: ({df1}, {df2})")
    print(f"P-Value: {p_value:.4f}")
    print(f"Critical F-Values: ({F_critical_low:.4f}, {F_critical_high:.4f})")

    if F < F_critical_low or F > F_critical_high:
        print("\nReject the null hypothesis (H0): The variances are significantly different.")
    else:
        print("\nFail to reject the null hypothesis (H0): No significant difference in variances.")

    plt.figure(figsize=(8, 5))
    sns.kdeplot(sample1, label="Sample 1", fill=True, color="blue", alpha=0.5)
    sns.kdeplot(sample2, label="Sample 2", fill=True, color="red", alpha=0.5)
    plt.axvline(np.mean(sample1), color='blue', linestyle='--', label=f"Mean 1: {np.mean(sample1):.2f}")
    plt.axvline(np.mean(sample2), color='red', linestyle='--', label=f"Mean 2: {np.mean(sample2):.2f}")
    plt.title("Density Plot of Two Samples")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

f_test_variance(sample1, sample2)
```

```python
#Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
'''
ANSWER
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def chi_square_goodness_of_fit(observed, expected, alpha=0.05):
    chi_square_stat = np.sum((np.array(observed) - np.array(expected))**2 / np.array(expected))

    df = len(observed) - 1

    p_value = 1 - stats.chi2.cdf(chi_square_stat, df)

    chi_critical = stats.chi2.ppf(1 - alpha, df)

    print("\n**Chi-Square Goodness of Fit Test:**")
    print(f"Observed Frequencies: {observed}")
    print(f"Expected Frequencies: {expected}")
    print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-Value: {p_value:.4f}")
    print(f"Critical Chi-Square Value (α={alpha}): {chi_critical:.4f}")

    if chi_square_stat > chi_critical:
        print("\nReject the null hypothesis (H0): The observed distribution is significantly different.")
    else:
        print("\nFail to reject the null hypothesis (H0): No significant difference in distribution.")

    categories = range(len(observed))
    plt.figure(figsize=(8, 5))
    plt.bar(categories, observed, label="Observed", color="blue", alpha=0.6)
    plt.bar(categories, expected, label="Expected", color="red", alpha=0.6)
    plt.xticks(categories)
    plt.xlabel("Categories")
    plt.ylabel("Frequency")
    plt.title("Chi-Square Goodness of Fit: Observed vs. Expected")
    plt.legend()
    plt.grid()
    plt.show()

np.random.seed(42)
observed_data = np.random.randint(20, 40, size=6)
expected_data = [sum(observed_data) / 6] * 6

chi_square_goodness_of_fit(observed_data, expected_data)
```
